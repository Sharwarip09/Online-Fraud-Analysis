# Model Fitting

```python
X = data.drop(columns=['isFraud','newbalanceOrig','newbalanceDest','nameDest','nameOrig'], axis=1)
print(X)
```

```python
from sklearn.model_selection import train_test_split
x = X
y = data['isFraud']
```

```python
train_x, test_x, train_y, test_y = train_test_split(x, y, test_size=0.3, random_state=42, stratify=y)
```

```python
import warnings
warnings.filterwarnings("ignore")
```

# LOGISTIC REGRESSION
Logistic Regression model gives accuracy as 99%.
Yet the model is not suitable as the precision and recall of positive instances is good but for negative instances model is unable to predict acurately.

```python
from sklearn.linear_model import LogisticRegression
from sklearn import metrics
from sklearn.metrics import classification_report
model1=LogisticRegression()
model1.fit(x,y)
predict1=model1.predict(test_x)
```

```python
predicted_isFraud=pd.DataFrame(predict1)
predicted_isFraud.value_counts()
```

```python
from sklearn.metrics import confusion_matrix
```

```python
metrics.accuracy_score(predicted_isFraud, test_y)
```

```python
print(classification_report(test_y, predict1))
```

```python
confusion_mat=confusion_matrix(test_y,predict1)
sns.heatmap(confusion_mat, annot=True, fmt='d', cmap='Blues',
            xticklabels=['Predicted 0', 'Predicted 1'],
            yticklabels=['Actual 0', 'Actual 1'])
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title('Confusion Matrix')
```

# RANDOMFOREST CLASSIFIER
RandomForest model gives accuracy as 99%.
The model is somewhat suitable as the precision and recall of positive instances is good but for negative instances precision and recall is not as good.

```python
from sklearn.ensemble import RandomForestClassifier
model2= RandomForestClassifier(random_state=1, n_estimators = 50)
model2.fit(train_x, train_y)
predict2 = model2.predict(test_x)
predict2
predict2=pd.DataFrame(predict2)
predict2.value_counts()
```

```python
metrics.accuracy_score(predict2, test_y)
```

```python
print(classification_report(test_y, predict2))
```

```python
confusion_mat=confusion_matrix(test_y,predict2)
sns.heatmap(confusion_mat, annot=True, fmt='d', cmap='Blues', 
            xticklabels=['Predicted 0', 'Predicted 1'],
            yticklabels=['Actual 0', 'Actual 1'])
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title('Confusion Matrix')
```

```python
from sklearn.metrics import roc_curve, roc_auc_score, auc
y_probs_rf = model2.predict_proba(test_x)[:, 1]
# Calculate ROC curve and AUC
fpr_rf, tpr_rf, thresholds_rf = roc_curve(test_y, y_probs_rf)
roc_auc_rf = auc(fpr_rf, tpr_rf)
# Plot ROC curve
plt.plot(fpr_rf, tpr_rf, color='darkorange', lw=2, label=f'ROC curve (AUC = {roc_auc_rf:.2f})')
plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--', label='Random Guess')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic (ROC) Curve - RandomForestClassifier')
plt.legend(loc='lower right')
```

# DEEP LEARNING ALGORITHM

```python
import tensorflow as tf 
from tensorflow import keras
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Flatten,Dense,InputLayer
from tensorflow.keras.optimizers import Adam  
```

```python
input_shape=X.shape[1:]
input_shape
```

```python
class_weight=y.value_counts().to_dict()
class_weight
```

```python
total_sample=sum(class_weight.values())
class_weight=[(total_sample)/(len(class_weight)*(class_weight[0])),(total_sample)/(len(class_weight)*(class_weight[1]))]
```

```python
print(class_weight)
```

```python
class_weight={0:0.5006462444095885,1:387.35054182393765}
class_weight
```

```python
model=Sequential()
model.add(InputLayer(input_shape))
model.add(Dense(128,activation='relu'))
model.add(Dense(128,activation='relu'))
model.add(Dense(1,activation='sigmoid'))
```

```python
model.summary()
'''Model has three layers dense,dense_1 and dense_2
dense and dense_1 layer has 128 ouputs where as last dense_2 has 1 outputs
 '''
```

```python
##compling the model
model.compile(optimizer=Adam(learning_rate=0.001),loss='binary_crossentropy',metrics=['accuracy'])
```

```python
history=model.fit(train_x,train_y,batch_size=32,epochs=10,class_weight=class_weight,validation_split=0.1)
```

```python
hist=pd.DataFrame(history.history)
hist.plot()
```
